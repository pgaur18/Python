## Plotting a function x^3

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([1,2,3,4])

y= x**3

plt.plot(x, y, "r--")
plt.show()
```

### The above graph is not that consistent because the values a not continuous 

### To generate continuous values we will use arange method 

```python
a= np.arange(0, 5, 0.2)

b= a**3

plt.plot(a, b, "g-")
plt.show()
```

## Only 1 argument

### The plot method works even if we don't pass the x argument to it. 
### It considers the given argument as "y" and x values are automatically taken in the range using 
### arange method -->  np.arange(0, n , step ) . This gives no between 0 to n-1 with the stepvalue as step

```python
# the array of values passed are considered to be as y points and x points are automatically taken as x= np.arange(0, n, 1)
x= np.array([3,4,5])

plt.plot(x, "b--")
plt.show()
```
