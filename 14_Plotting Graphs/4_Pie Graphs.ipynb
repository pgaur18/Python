```python
import matplotlib.pyplot as plt
```

```python
# sizes or division in pie graphs

sizes= [3,4,6,2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
# Adding diffeerent colour to the pie graph

sizes= [3,4,6,2]
colors= ["yellow", "blue", "red", "purple"]

plt.pie(sizes, colors= colors)
plt.axis("equal")
plt.show()
```

```python
# Adding labels ans Title

sizes= [3,4,6,2]
colors= ["yellow", "blue", "red", "purple"]

labels= ["A", "B", "C", "D"]

plt.title("Pie Chart Demo")

plt.pie(sizes, colors= colors, labels= labels)
plt.axis("equal")

plt.legend() 

plt.show()
```

### autopct takes formated string with the syntax
### autopct= " %0.2f %% "

#### 2 indicates how many digits we want after decimal ( that is why . dot is used --> dot means decimal )

```python
# Displaying the percentage 

sizes= [ 2, 2, 2, 2 ]
colors= ["yellow", "blue", "red", "purple"]

labels= ["A", "B", "C", "D"]

plt.title("Pie Chart Demo")

plt.pie(sizes, colors= colors, labels= labels, autopct= "%.2f%%")
plt.axis("equal")

plt.legend()
plt.show()
```

```python
# Explode or Highlight a particular section

sizes= [3,4,6,2]
labels= [ "A", "B", "C", "D" ]
colors= [ "yellow", "red", "pink", "blue" ]

explode= [.2, 0, 0, 0]

plt.title("Pie Chart Demo")
plt.axis("equal")
plt.pie(sizes, colors= colors, labels= labels, explode= explode, autopct= " %.2f%% ")

plt.legend()
plt.show()
```

```python
# Counter Clockwise

sizes= [3,4,6,2]
labels= ["A", "B", "C", "D"]
colors= ["yellow", "red", "pink", "blue"]

explode= [.2, 0, 0, 0]
plt.title("Pie Chart Demo")
plt.axis("equal")
plt.pie(sizes, colors= colors, labels= labels, counterclock= False ,explode= explode, autopct= "%.2f%%")
plt.legend()
plt.show()
```

```python
# Change the start point or angle

sizes= [3,4,6,2]
labels= ["A", "B", "C", "D"]
colors= ["yellow", "red", "pink", "blue"]

explode= [.2, 0, 0, 0]
plt.title("Pie Chart Demo")
plt.axis("equal")
plt.pie(sizes, colors= colors, labels= labels, explode= explode, counterclock= False , autopct= "%.2f%%", startangle= 90)
plt.show()
```
