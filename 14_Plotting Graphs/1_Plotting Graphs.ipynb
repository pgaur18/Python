```python
import matplotlib.pyplot as plt
```

```python
# Only points are visible

x= [1,2,3]
y= [2,4,6]

plt.scatter(x,y)

plt.show()
```

```python
# Connect the points with a line. Points are not visible

x= [1,2,3]
y= [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
# Both the points and line are visible. Line passes through the points

x= [1,2,3]
y= [2,4,6]

plt.scatter(x,y)
plt.plot(x,y)

plt.show()
```

```python
# Plot more than one line

x1= [1,2,3]
y1= [2,4,6]

x2= [3,4,5]
y2= [2,4,6]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()
```

```python
# Lines with colour

x1= [1,2,3]
y1= [2,4,6]

x2= [3,4,5]
y2= [2,4,6]

plt.plot(x1,y1, "r-")
plt.plot(x2,y2, "y-")

plt.show()
```

```python
x1= [1,2,3]
y1= [2,4,6]

x2= [3,4,5]
y2= [2,4,6]

# p= pentagon

# plt.plot(x1, y1, "g*")
# plt.plot(x2, y2, "y--")

plt.plot(x1, y1, "g--")
plt.plot(x2, y2, "y-.")

plt.show()
```

```python
x1= [1,2,3]
y1= [2,4,6]

x2= [3,4,5]
y2= [2,4,6]

x3= [1,4,6]
y3= [6,2,8]

plt.plot(x1,y1, "r-")
plt.plot(x2,y2, "y-")
plt.plot(x3,y3, "o-")

plt.show()
```
