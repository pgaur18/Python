```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(0, 5, 0.2)

y= x**3
plt.plot(y, "y--")
plt.show()
```

```python
# Different way of colour and marker

x= np.arange(0, 5, 0.2)

# colour is wrong
# color is correct

y= x**3
plt.plot(y, color= "blue", marker= "*")
plt.show()
```

```python
# Changing the line width

x= np.arange(0, 5, 0.2)
y= x**3
plt.plot(y, "y--", linewidth= 4)

plt.show()
```

```python
# X and Y axis labels and Title of the graphs

x= np.arange(0, 5, 0.2)
y= x**3
plt.plot(y, "y--", linewidth= 4)

plt.title("Matplotlib.pyplot Demo")
plt.ylabel("X^3")
plt.xlabel("X")

plt.show()
```

```python
# Naming the lines in graph. Helps in distinguishing between the lines just by seeing
x1= np.arange(0, 5, 0.2)
y1= x**3

y2= x**2

plt.plot(y1, "y--", linewidth= 2, label= "X^3")
plt.plot(y2, "r-", linewidth= 2, label= "x^2")
plt.legend()


plt.title("Matplotlib.pyplot Demo")
plt.ylabel("X^3")
plt.xlabel("X")

plt.show()
```

```python
# Changing the axis values and grid

x1= np.arange(0, 5, 0.1)
y1= x**3

y2= x**2


plt.plot(y1, "y--", linewidth= 2, label= "X^3")
plt.plot(y2, "r-", linewidth= 2, label= "x^2")

plt.axis([0, 10, 0, 10])

plt.legend()

plt.grid()
plt.title("Matplotlib.pyplot Demo")
plt.ylabel("X^3")
plt.xlabel("X")

plt.show()
```

```python
# Adding text

x1= np.arange(0, 5, 0.1)
y1= x**3

y2= x**2


plt.plot(y1, "y--", linewidth= 2, label= "X^3")
plt.plot(y2, "r-", linewidth= 2, label= "x^2")

plt.axis([0, 10, 0, 10])

plt.legend()

plt.text(4, 4, "Coordinate (4,4) ", fontsize= 15)

plt.grid()
plt.title("Matplotlib.pyplot Demo")
plt.ylabel("X^3")
plt.xlabel("X")

plt.show()
```
