```python
import pandas as pd

iris= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")

iris
```

```python
df= iris.copy()

df.columns= ["SL", "SW", "PL", "PW", "Flower_Type"]

df.head()
```

```python
df["Gender"]= "Female"

df.head()
```

```python
df.iloc[0:10, 5]= "Male"

df.head()

df.tail()
```

```python
df.head()
```

```python
def f( str ):
    if( str == "Male" ):
        return 1
    else:
        return 0
```

```python
df["Sex"]= df.Gender.apply(f)
```
