```python
df= iris.copy()

df.columns= ["SL", "SW", "PL", "PW", "Flower_Type"]

df.head()
```

# How to handle NAN values ? 
## Way 1

### Drop the Nan entries

```python
df.dropna(inplace= True)

df
```

```python
df.reset_index(drop= True, inplace= True)

df.head()
```

## Assign some value to NaN entries

```python
df.iloc[2:4, 1:3]= np.nan

df.head()
```

```python
df.SW.fillna(df.SW.mean(), inplace= True)

df.PL.fillna(df.PL.mean(), inplace= True)
```

```python
df.head()
```

## Filling it with mean of those type of specific flowers

```python
# a is a data frame variable. It is same as df
a= df[df.Flower_Type == "Iris-setosa"]

df.SW.fillna(a.SW.mean(), inplace= True)

df.head()


b= df[df.Flower_Type == "Iris-setosa"]

df.PL.fillna(b.PL.mean(), inplace= True)

df.head()
```
